## Step 1: Importing Libraries

```python
import numpy as np
import pandas as pd
```

## Step 2: Loading the data and selecting the feature columns

```python
df = pd.read_csv('../data/raw/developer-survey-2025.csv', low_memory=False)
```

```python
# Define the exact columns we want to keep.
# This is a Python LIST of column name strings.
# We are deliberately ignoring the other 162 columns.

cols_to_keep = [
    'ConvertedCompYearly',  # TARGET - salary we want to predict
    'Country',              # Where the developer works
    'YearsCode',            # Total years of coding experience
    'WorkExp',              # Years of PROFESSIONAL work experience
    'EdLevel',              # Highest education level
    'DevType',              # Type of developer role
    'OrgSize',              # Size of their employer
    'RemoteWork',           # Remote / hybrid / in-person
    'Employment',           # Employment status
    'Age'                   # Age bracket
]

# df[cols_to_keep] selects ONLY those columns from the full DataFrame.
# We immediately overwrite df with this smaller version.
# This makes every subsequent step faster and cleaner.

df = df[cols_to_keep]

print(f'Columns kept: {df.shape[1]}')
print(f'Total rows:   {df.shape[0]:,}')
df.head()
```

## Step 3: Drop rows without Salary (Target column)

```python
# We can only train a model when we KNOW the answer (the salary).
# Any row where ConvertedCompYearly is missing is useless for training.
#
# Before dropping, let's record how many rows we're losing

before = len(df)

# .dropna(subset=['ConvertedCompYearly']) removes rows where
# ConvertedCompYearly specifically is NaN.
# subset= means "only drop if THIS column is missing"
# Without subset=, it would drop rows where ANY column is missing — too aggressive.
#
# inplace=True modifies df directly instead of returning a new DataFrame.
# It's equivalent to writing: df = df.dropna(subset=['ConvertedCompYearly'])

df.dropna(subset=['ConvertedCompYearly'], inplace=True)

after = len(df)

print(f'Rows before removing missing salary: {before:,}')
print(f'Rows after:                          {after:,}')
print(f'Rows dropped:                        {before - after:,}')
print(f'\nWe now have {after:,} rows to work with.')
```
